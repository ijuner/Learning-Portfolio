2311. Longest Binary Subsequence Less Than or Equal to K
You are given a binary string s and a positive integer k.
Return the length of the longest subsequence of s that makes up a binary number less than or equal to k.

```python
def longestSubsequence(s: str, k: int) -> int:
    n = len(s)
    val = 0
    power = 1
    cnt = 0

    # 从后往前枚举（即从低位开始）
    for i in range(n - 1, -1, -1):
        if s[i] == '1':
            if power <= k and val + power <= k:
                val += power
                cnt += 1
            # 提前终止：如果再乘2已经超过K，就不可能再加更多1了
        else:  # s[i] == '0'
            cnt += 1  # '0'不会增加val，无论如何都能加

        # 更新power，如果下次会超范围就停止处理（防止 power 溢出）
        if power * 2 > k:
            break
        power *= 2

    # 加上剩余的0（在没处理到的位中）
    for j in range(i - 1, -1, -1):
        if s[j] == '0':
            cnt += 1

    return cnt
```

```python
s = "1001010"
k = 5
print(longestSubsequence(s,k))

s1 = "00101001"
k1 = 1
print(longestSubsequence(s1,k1))
```
